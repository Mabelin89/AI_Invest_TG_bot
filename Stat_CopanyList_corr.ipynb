```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from moex_parser import get_historical_data
from datetime import datetime

# Фильтрация только обыкновенных акций
tickers = df[df['stock_type'] == 'Обыкновенная']['ticker'].unique()

# Параметры
timeframe = 'weekly'
start_date = '2020-04-27'
end_date = datetime.now().strftime('%Y-%m-%d')

# Словарь для хранения данных
data_dict = {}

# Парсинг данных для каждого тикера
for ticker in tickers:
    print(f"Загрузка данных для {ticker}...")
    data = get_historical_data(ticker, timeframe, period_years=5)  # period_years игнорируется в moex_parser, но оставлен для совместимости
    
    # Проверка на наличие данных
    if data is None or data.empty or len(data) == 0:
        print(f"Нет данных для {ticker}")
        continue
    
    # Фильтрация данных с 2020-04-27
    data['date'] = pd.to_datetime(data['date'])
    data = data[data['date'] >= start_date]
    
    # Проверка на наличие данных за период
    if data.empty:
        print(f"Нет данных для {ticker} после {start_date}")
        continue
    
    # Проверка на достаточный период (примерно 5 лет)
    data_time_span = (data['date'].max() - data['date'].min()).days
    if data_time_span < 5 * 365 - 30:  # Допуск 30 дней
        print(f"Недостаточно данных для {ticker} (менее 5 лет с {start_date})")
        continue
    
    # Сохранение цен закрытия
    data_dict[ticker] = data.set_index('date')['close']

# Создание DataFrame с ценами закрытия
prices_df = pd.DataFrame(data_dict)

# Проверка на наличие данных
if prices_df.empty:
    print("Нет данных для построения корреляции")
else:
    # Расчёт корреляционной матрицы
    corr_matrix = prices_df.corr()

    # Выбор топ-15 пар с наибольшей корреляцией
    corr_unstack = corr_matrix.unstack()
    corr_unstack = corr_unstack[corr_unstack < 1.0]  # Исключаем корреляцию с самим собой
    top_15_pairs = corr_unstack.abs().sort_values(ascending=False).head(15)

    print("\nТоп-15 пар с наибольшей корреляцией:")
    print(top_15_pairs)

    # Выбор тикеров для тепловой карты (уникальные из топ-15 пар)
    top_tickers = set()
    for (ticker1, ticker2) in top_15_pairs.index:
        top_tickers.add(ticker1)
        top_tickers.add(ticker2)
    top_tickers = list(top_tickers)

    # Ограничение корреляционной матрицы для топ-тикеров
    top_corr_matrix = corr_matrix.loc[top_tickers, top_tickers]

    # Построение тепловой карты
    plt.figure(figsize=(10, 8))
    sns.heatmap(top_corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title('Тепловая карта корреляции цен закрытия (Топ-15 пар, обыкновенные акции, с 2020-04-27)')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from moex_parser import get_historical_data
from itertools import combinations

# Предполагается, что df уже загружен и содержит столбец 'ticker'
tickers = df['ticker'].unique().tolist()
print(f"Всего тикеров: {len(tickers)}")

# Параметры
timeframe = 'weekly'
period_years = 7

# Словарь для хранения данных
weekly_data = {}

# Получение данных для каждого тикера
for ticker in tickers:
    print(f"Загрузка данных для {ticker}...")
    data = get_historical_data(ticker, timeframe, period_years)
    if not data.empty:
        data['date'] = pd.to_datetime(data['date'])
        data.set_index('date', inplace=True)
        weekly_data[ticker] = data['close']
    else:
        print(f"Данные для {ticker} не получены.")

# Объединяем данные в один DataFrame
if weekly_data:
    combined_data = pd.DataFrame(weekly_data)
    print(f"Объединённый DataFrame: {combined_data.shape}")

    # Удаляем столбцы с большим количеством пропусков (>50%)
    combined_data = combined_data.dropna(thresh=len(combined_data)*0.5, axis=1)
    print(f"После удаления столбцов с пропусками: {combined_data.shape}")

    # Заполняем пропуски
    combined_data = combined_data.fillna(method='ffill').fillna(method='bfill')
    
    # Вычисляем корреляционную матрицу
    corr_matrix = combined_data.corr(method='pearson')

    # Извлекаем пары корреляций
    corr_pairs = []
    for ticker1, ticker2 in combinations(corr_matrix.columns, 2):
        corr_value = corr_matrix.loc[ticker1, ticker2]
        corr_pairs.append((ticker1, ticker2, corr_value))

    # Создаём DataFrame из пар корреляций
    corr_df = pd.DataFrame(corr_pairs, columns=['Ticker1', 'Ticker2', 'Correlation'])
    
    # Топ-15 пар с положительной корреляцией
    top_positive = corr_df[corr_df['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(15)
    top_positive['Pair'] = top_positive['Ticker1'] + ' - ' + top_positive['Ticker2']
    
    # Топ-15 пар с отрицательной корреляцией
    top_negative = corr_df[corr_df['Correlation'] < 0].sort_values(by='Correlation', ascending=True).head(15)
    top_negative['Pair'] = top_negative['Ticker1'] + ' - ' + top_negative['Ticker2']
    
    # Получаем уникальные тикеры из топ-15 положительных и отрицательных пар
    positive_tickers = list(set(top_positive['Ticker1']).union(set(top_positive['Ticker2'])))
    negative_tickers = list(set(top_negative['Ticker1']).union(set(top_negative['Ticker2'])))
    all_top_tickers = list(set(positive_tickers).union(set(negative_tickers)))
    print(f"Тикеры для тепловой карты: {all_top_tickers}")

    # Фильтруем корреляционную матрицу для топ-тикеров
    top_corr_matrix = corr_matrix.loc[all_top_tickers, all_top_tickers].copy()

    # Создаём маски для положительных и отрицательных корреляций
    positive_mask = np.zeros_like(top_corr_matrix, dtype=bool)
    negative_mask = np.zeros_like(top_corr_matrix, dtype=bool)
    
    # Заполняем маски для топ-15 положительных и отрицательных пар
    for _, row in top_positive.iterrows():
        i = top_corr_matrix.index.get_loc(row['Ticker1'])
        j = top_corr_matrix.columns.get_loc(row['Ticker2'])
        positive_mask[i, j] = True
        positive_mask[j, i] = True  # Симметрия
    
    for _, row in top_negative.iterrows():
        i = top_corr_matrix.index.get_loc(row['Ticker1'])
        j = top_corr_matrix.columns.get_loc(row['Ticker2'])
        negative_mask[i, j] = True
        negative_mask[j, i] = True  # Симметрия
    
    # Создаём новую матрицу: верхний треугольник для положительных, нижний для отрицательных
    final_matrix = np.zeros_like(top_corr_matrix)
    upper_triangle = np.triu_indices_from(final_matrix, k=1)  # Выше диагонали
    lower_triangle = np.tril_indices_from(final_matrix, k=-1)  # Ниже диагонали
    
    # Заполняем верхний треугольник положительными корреляциями
    final_matrix[upper_triangle] = np.where(positive_mask[upper_triangle], top_corr_matrix.values[upper_triangle], 0)
    
    # Заполняем нижний треугольник отрицательными корреляциями
    final_matrix[lower_triangle] = np.where(negative_mask[lower_triangle], top_corr_matrix.values[lower_triangle], 0)
    
    # Преобразуем в DataFrame для тепловой карты
    final_matrix_df = pd.DataFrame(final_matrix, index=top_corr_matrix.index, columns=top_corr_matrix.columns)
    
    # Построение тепловой карты
    plt.figure(figsize=(12, 10))
    sns.heatmap(final_matrix_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f', mask=final_matrix_df == 0)
    plt.title('Тепловая карта: Топ-15 положительных (выше диагонали) и отрицательных (ниже диагонали) корреляций')
    plt.tight_layout()
    plt.show()
    
    # Вывод таблиц
    print("\nТоп-15 пар с положительной корреляцией:")
    print(top_positive[['Pair', 'Correlation']].to_string(index=False))
    
    print("\nТоп-15 пар с отрицательной корреляцией:")
    print(top_negative[['Pair', 'Correlation']].to_string(index=False))
else:
    print("Нет данных для анализа корреляции.")
```
